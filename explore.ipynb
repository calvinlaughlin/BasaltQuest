```python
import os
import geopandas as gpd

directory = 'BasaltQuest_data'

shapefiles = {}

for root, dirs, files in os.walk(directory):
    for file in files:
        if file.endswith('.shp'):
            # Construct full file path
            file_path = os.path.join(root, file)
            # Add to dictionary with filename (without extension) as key
            shapefiles[os.path.splitext(file)[0]] = file_path

# Print the dictionary of shapefiles
shapefiles
```

```python
geo_dfs = {}

for key, path in shapefiles.items():
    geo_dfs[key] = gpd.read_file(path)
    print(f"['{key}'] data loaded successfully.")
print('------------------------------------')
geo_dfs
```

```python
import geopandas as gpd
import folium

brazil_gdf = geo_dfs['BR_Pais_2022']  # Brazil country boundary
# mines and ag
mines_gdf = geo_dfs['Basalt mines — licences_basalt_mines']
agri_gdf = geo_dfs['IBGE22_Annual_crop_production']
# transportation
hidrovias_gdf = geo_dfs['hidrovias_linha_2014']
rodovia_gdf = geo_dfs['rodovia_2014']
# basalt dissolved
basalt_25km_gdf = geo_dfs['basalt_Dissolved_25km']
basalt_50km_gdf = geo_dfs['basalt_Dissolved_50km']
basalt_75km_gdf = geo_dfs['basalt_Dissolved_75km']
basalt_100km_gdf = geo_dfs['basalt_Dissolved_100km']
basalt_150km_gdf = geo_dfs['basalt_Dissolved_150km']

# Initialize the map centered on Brazil
m = folium.Map(location=[-14.2350, -51.9253], zoom_start=4)

def add_geojson(gdf, name, color):
    folium.GeoJson(
        gdf,
        name=name,
        style_function=lambda feature: {
            'fillColor': color,
            'color': color,
            'weight': 2,
            'dashArray': '5, 5'
        }
    ).add_to(m)

# Add Brazil boundary
add_geojson(brazil_gdf, 'Brazil', 'gray')

# Add basalt mines
add_geojson(mines_gdf, 'Basalt Mines w/ Licenses', 'red')

# Add agricultural centers
# add_geojson(agri_gdf, 'Annual Crop Production', 'green')

# Add proximity layers
add_geojson(basalt_25km_gdf, '25km Buffer', 'blue')
add_geojson(basalt_50km_gdf, '50km Buffer', 'yellow')
add_geojson(basalt_75km_gdf, '75km Buffer', 'purple')
add_geojson(basalt_100km_gdf, '100km Buffer', 'orange')
add_geojson(basalt_150km_gdf, '150km Buffer', 'brown')

folium.LayerControl().add_to(m)
m.save('basalt_mines_and_agricultural_centers.html')
print("Map saved as 'basalt_mines_and_agricultural_centers.html'")
```

```python
mines_gdf = geo_dfs['Basalt mines — licences_basalt_mines']

# Reproject to a projected CRS (UTM Zone 23S)
mines_gdf = mines_gdf.to_crs(epsg=32723)

coordinates = mines_gdf['geometry'].centroid

# Convert centroids to a geographic CRS for visualization in Folium (which uses geographic CRS)
coordinates = coordinates.to_crs(epsg=4326)

coordinates
```

```python
import folium

# Initialize a Folium map centered on Brazil
map_center = [coordinates.y.mean(), coordinates.x.mean()]
m = folium.Map(location=map_center, zoom_start=5)

# Add markers for a few coordinates
for coord in coordinates[:10]:  # Limit to first 10 for simplicity
    folium.Marker(location=[coord.y, coord.x]).add_to(m)

# Save the map to an HTML file and display
m.save("map.html")

# Display the map in the notebook (this will only work in a Jupyter environment)
m
```

```python
# Import required libraries
import ee
import os
import requests
from tqdm import tqdm
import geopandas as gpd

# Authenticate and initialize the Earth Engine API
ee.Authenticate()
ee.Initialize()

# Define visualization parameters
vis_params = {
    'bands': ['B4', 'B3', 'B2'],  # True color bands
    'min': 0,
    'max': 3000,
    'gamma': 1.4,
}

# Initialize a list to store URLs
image_urls = []

# Create a directory to save images if it doesn't exist
output_dir = "satellite_images"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Function to download images
def download_image(url, path):
    response = requests.get(url)
    with open(path, 'wb') as file:
        file.write(response.content)

# Loop through each coordinate and download the corresponding image
for index, point in tqdm(enumerate(coordinates), total=len(coordinates), desc="Downloading Images"):
    coords = ee.Geometry.Point([point.x, point.y])
    
    # Load a satellite image collection (Sentinel-2)
    collection = ee.ImageCollection('COPERNICUS/S2') \
                  .filterBounds(coords) \
                  .filterDate('2020-01-01', '2020-12-31') \
                  .sort('CLOUDY_PIXEL_PERCENTAGE') \
                  .first()
    
    # Define a 5000m by 5000m region around the point
    region = coords.buffer(1000).bounds().getInfo()['coordinates']
    
    # Select the specific bands for true color visualization
    image = collection.select(['B4', 'B3', 'B2']).clip(coords.buffer(1000))
    
    # Get the image URL
    url = image.getThumbURL({
        'min': 0,
        'max': 3000,
        'dimensions': '1024x1024',
        'region': region
    })
    
    # Add the URL to the list
    image_urls.append(url)
    
    # Download the image
    download_image(url, os.path.join(output_dir, f'sentinel_image_{index}.png'))

# Display the list of downloaded image URLs
print("Downloaded images.")
# for url in image_urls:
#     print(url)
```
